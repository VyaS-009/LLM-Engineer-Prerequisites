## 🧠 Concept: Linear Regression

Linear Regression predicts a continuous value:

$$y = w x + b$$

- $w$: slope (weight)
- $b$: intercept (bias)

**Loss Function (Mean Squared Error):**

$$L = \frac{1}{N} \sum_i (y_i - (w x_i + b))^2$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = np.linspace(0, 10, 50)
y = 2.5 * X + 5 + np.random.randn(50) * 2

# Initialize parameters
w, b = np.random.randn(), np.random.randn()
lr = 0.01
epochs = 200
losses = []

for epoch in range(epochs):
    y_pred = w * X + b
    error = y_pred - y
    
    # Compute gradients
    dw = (2 / len(X)) * np.dot(error, X)
    db = (2 / len(X)) * np.sum(error)
    
    # Update weights
    w -= lr * dw
    b -= lr * db
    
    loss = np.mean(error ** 2)
    losses.append(loss)

plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, w * X + b, color='red', label=f'Fitted Line: y={w:.2f}x+{b:.2f}')
plt.legend()
plt.title('Linear Regression Fit via Gradient Descent')
plt.show()
```

## ⚙️ Gradient Descent Intuition

We want to find parameters $(w, b)$ that minimize $L(w, b)$.

Update rule:

$$w := w - \alpha \frac{\partial L}{\partial w}$$
$$b := b - \alpha \frac{\partial L}{\partial b}$$

- $\alpha$: learning rate (step size)
- Gradients tell us *which direction* to move to reduce loss.

```python
# Plot the loss convergence
plt.plot(range(epochs), losses)
plt.title('Loss Convergence over Epochs')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.show()
```

## 🏁 Real-World Analogy
Think of Gradient Descent like **walking down a hill blindfolded**:

- The slope (gradient) tells you which direction to step.
- The learning rate decides how *big* your steps are.
- You stop when the ground feels flat — i.e., loss stops improving.

💡 This same principle drives how deep neural networks learn millions of parameters.
